```python
df = pd.read_csv('titanic_data.csv')
df.head()
df.columns
```

```python
for i in range(len(df.columns[])):
  label = df.columns[i]
  plt.hist(df[df['Survived']==1][label], color='blue', label="Survived", alpha=0.7, density=True, bins=15)
  plt.hist(df[df['Survived']==0][label], color='red', label="Deceased", alpha=0.7, density=True, bins=15)
  plt.title(label)
  plt.ylabel("N")
  plt.xlabel(label)
  plt.legend()
  plt.show()
```
